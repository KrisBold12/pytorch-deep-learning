### Matrix multiplication
Two main ways:
1. Element-wise multiplication
2. Matrix multiplication

Two main rules:
1. The **inner dimensions** must match:
* (3, 2) @ (3, 2) won't work
* (2, 3) @ (3, 2) will work
* (3, 2) @ (2, 3) will work
2. The resulting matrix has the shape of the **outer dimensions**
* (2, 3) @ (3, 2) -> (2, 2)
* (3, 2) @ (2, 3) -> (3, 3)

```python
# Element-wise
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
%%time
torch.matmul(tensor, tensor)
```

## Reshaping, stacking, squeezing and unsqueezing tensors


```python
x = torch.arange(1, 10)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
x = torch.arange(1, 13)
x, x.shape
```

```python
x_reshaped = x.reshape(3, 2, 2)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(3, 4)
z, z.shape
```

```python
# Changing z changes x, because the view of a tensor shares the same memory address as the original input
z[:, 0] = 5
z, x
```

```python
# Stack tensor on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# torch.squeeze() - removes all single dimensions from a tensor
x = torch.zeros(2, 1, 2, 1, 2)
print(x)
x.size()
```

```python
y = torch.squeeze(x)
print(y)
y.size()
```

```python
y = torch.squeeze(x, 0)
print(y)
y.size()
```

```python
y = torch.squeeze(x, 1)
print(y)
y.size()
```

```python
y = torch.squeeze(x, 3)
print(y)
y.size()
```

```python
# torch.unsqueeze() - adds a single dimension to a tensor at a specific dim
x_squeezed = x_reshaped.squeeze()
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimension with unsqueeze
y = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {y}")
print(f"New shape: {y.shape}")
```

```python
# torch.permute() - rearranges the dimensions of a target tensor - note: it return a view (same mem address)
x_original = torch.rand(size=(224, 224, 3)) # height, width, colour_channels

# Permute the original tensor do rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0 -> 1, 1 -> 2, 2 -> 0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # colour_channel, height, width
```

## PyTorch tensors and NumPy
* Data in NumPy, want in PyTorch tensor -> torch.from_numpy(ndarray)
* PyTorch tensor -> NumPy -> torch.Tensor.numpy()


```python
# NumPy to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: when converting from numpy -> pytorch, pytorch reflects numpy's default dtype of float64 unless specified otherwise
array, tensor
```

```python
# Change the value of the array, what happens to tensor
array += 1
array, tensor
```

```python
# Tensor to numpy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

## Reproducibility
How neural network learns:
Start with random numbers -> tensor operations -> update random numbers to try and make them better represent the data -> repeat

To reduce the randomness in neural networks and pytorch comes the concept of a **random seed**: 'flavour' the randomness

```python
import torch

random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A == random_tensor_B)
```

```python
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED) # Need to call again
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C == random_tensor_D)
```
